### QUESTION 1:

#### Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

##### Seaborn is a powerful Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics. Here are five commonly used plots in Seaborn along with their typical uses:

##### 1. **Histogram (`sns.histplot`):**
#####   - **Use:** Visualize the distribution of a single numerical variable.
#####   - **Example:** Analyzing the distribution of exam scores in a class.

##### 2. **Scatter Plot (`sns.scatterplot`):**
#####   - **Use:** Show the relationship between two numerical variables.
#####   - **Example:** Investigating the correlation between study hours and test scores.

##### 3. **Bar Plot (`sns.barplot`):**
#####   - **Use:** Display the relationship between a categorical variable and a numerical variable.
#####   - **Example:** Comparing average sales revenue across different product categories.

##### 4. **Box Plot (`sns.boxplot`):**
#####   - **Use:** Summarize the distribution of a numerical variable across several categories.
#####   - **Example:** Analyzing the distribution of salaries across different job titles.

##### 5. **Heatmap (`sns.heatmap`):**
#####   - **Use:** Plot rectangular data as a color-encoded matrix, useful for visualizing correlations.
#####   - **Example:** Showing correlations between different features in a dataset.



### QUESTION 2:

####  Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = 
#### "timepoint" and y = "signal" for different events and regions. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
a=sns.load_dataset("fmri")
sns.lineplot(data=a,x='timepoint',y='signal',hue='event')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('FMRI Signal over Time by Event and Region')
plt.legend(title='Event')
plt.show()
```

### QUESTION 3:

#### Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'

```python
import seaborn as sns
data1=sns.load_dataset("titanic")
box1=sns.boxplot(data=data1,x='pclass',y='age')
plt.xlabel('pclass')
plt.ylabel('age')
plt.title('pclass vs age in titanic dataset')
plt.show()
```

```python
import seaborn as sns
data1=sns.load_dataset("titanic")
box2=sns.boxplot(data=data1,x='pclass',y='fare')
plt.xlabel('pclass')
plt.ylabel('fare')
plt.title('pclass vs fare in titanic dataset')
plt.show()
```

### QUESTION 4:

#### Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
a=sns.load_dataset("diamonds")
b=sns.histplot(data=a,x='price',hue='cut')
plt.xlabel('price')
plt.title('diamonds dataset')
plt.show()
```

### QUESTION 5

#### Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
import seaborn as sns
a=sns.load_dataset("iris")
sns.pairplot(a,hue='species')
plt.title('pairplot')
plt.show()
```

### QUESTION 6:

#### Use the "flights" dataset from seaborn to plot a heatmap

```python
import seaborn as sns
a=sns.load_dataset("flights")
flights_pivot = a.pivot(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('heatmap')
plt.show()
```
