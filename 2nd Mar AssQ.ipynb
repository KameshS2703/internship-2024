### QUESTION 2:

####  What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot

#### A scatter plot, also known as a scatter graph or scatter chart, is a type of plot or mathematical diagram that displays the relationship between two continuous variables. It uses Cartesian coordinates to display the values of two variables as points on a grid, with each point representing a single observation or data point.


```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("SCATTER PLOT")
```

### QUESTION 3:

#### Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:

#### The subplot() function is used to create multiple plots in a single figure. 

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])   
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0,0].plot(x1,y1)
axs[0,0].set_title("DATA1")
axs[0,1].plot(x2,y2)
axs[0,1].set_title("DATA2")
axs[1,0].plot(x3,y3)
axs[1,0].set_title("DATA3")
axs[1,1].plot(x4,y4)
axs[1,1].set_title("DATA4")
```

### QUESTION 4:

#### What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### A bar plot, also known as a bar chart or bar graph, is a graphical representation of data that uses bars to display the magnitude or frequency of different categories or groups. It is a type of categorical plot, where the x-axis represents the categories and the y-axis represents the values or frequencies.


```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

### QUESTION 5:

#### What is a box plot? Why is it used? Using the following data plot a box plot.

#### A box plot, also known as a box-and-whisker plot, is a graphical representation of data that displays the distribution of values in a dataset. It is a type of statistical plot that is used to summarize and visualize the central tendency, dispersion, and variability of a dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
```
