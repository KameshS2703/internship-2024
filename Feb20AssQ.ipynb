### ANSWER TO QUESTION 1:

####  Explain GET and POST methods.

```python
GET - Simply get is a method which is used to retrieve data from the server.
POST - Simply post is amethod which is sued to insert a data into a server.

GET:

##### - Retrieves data from the server
##### - Used for displaying or fetching data
##### - Request parameters are sent in the URL query string
##### - Data is retrieved from the server, but not modified

##### POST:

##### - Sends data to the server for processing
##### - Used for creating, inserting, or updating data
##### - Request parameters are sent in the request body
##### - Data is sent to the server, and the server processes and stores it

```

### ANSWER TO QUESTION 2:

####  Why is request used in Flask?

##### The request object is used in Flask to:

##### 1. Access request data: You can use request to access the data sent in the request, such as form data, query string parameters, or JSON data.
##### 2. Check request headers: You can use request to check the headers sent in the request, such as the Content-Type header.
##### 3. Get the requested URL: You can use request to get the URL that the user requested.
##### 4. Determine the HTTP method: You can use request to determine the HTTP method used in the request (GET, POST, etc.).



### ANSWER TO QUESTION 3:

#### Why is redirect() used in Flask?

##### In Flask, the redirect() function is used to redirect a user to a different URL. It's commonly used in the following scenarios:

##### 1. After form submission: After processing a form, you might want to redirect the user to a confirmation page or a new page.
##### 2. Authentication: After a user logs in, you might want to redirect them to a protected page or their dashboard.
##### 3. Error handling: If an error occurs, you might want to redirect the user to an error page or a page with more information.
##### 4. Route chaining: You might want to redirect the user to a different route or page after completing a specific task.



### ANSWER TO QUESTION 4:

#### What are templates in Flask? Why is the render_template() function used?

#### In Flask, templates are HTML files that contain placeholders for dynamic data. They allow you to separate the presentation layer (HTML) from the application logic (Python code).

#### The render_template() function is used to render a template and pass data from the Python code to the HTML template. It takes the template name and a dictionary of data as arguments, and returns a rendered HTML page with the data inserted into the placeholders.
#### HTML (user input) → Python (manipulation) → HTML (rendered with results)
