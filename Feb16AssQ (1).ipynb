### ANSWER TO QUESTION 1:

#### What is a database? Differentiate between SQL and NoSQL databases

##### A database is a collection of organized data that allows for efficient storage, retrieval, and manipulation of data.

##### SQL (Structured Query Language) databases, also known as relational databases, store data in tables with well-defined schemas. Each table has rows and columns, where each row represents a single record and each column represents a field or attribute of that record. SQL databases follow a strict schema, and data is stored in a normalized form to minimize redundancy.

##### Examples of SQL databases include:

##### - MySQL
##### - PostgreSQL
##### - Microsoft SQL Server
##### - Oracle

##### NoSQL (Not Only SQL) databases, on the other hand, store data in a variety of formats, such as key-value pairs, documents, graphs, or columns. NoSQL databases often have a dynamic schema, and data can be stored in a denormalized form. NoSQL databases are designed for scalability and flexibility, and are often used in big data and real-time web applications.

##### Examples of NoSQL databases include:

##### - MongoDB (document-oriented)
##### - Cassandra (column-oriented)
##### - Redis (key-value store)
##### - Neo4j (graph database)

##### Key differences between SQL and NoSQL databases:

##### - Schema: SQL databases have a fixed schema, while NoSQL databases have a dynamic schema.
##### - Data structure: SQL databases use tables, while NoSQL databases use various data structures like documents, graphs, or key-value pairs.
##### - Scalability: NoSQL databases are designed for horizontal scaling, while SQL databases can become bottlenecked as the dataset grows.
##### - ACID compliance: SQL databases follow ACID (Atomicity, Consistency, Isolation, Durability) principles, while NoSQL databases often sacrifice some of these principles for performance and scalability.

### What is DQL? Explain SELECT with an example.

#### DQL stands for Data Query Language, which is a subset of SQL (Structured Query Language) used to retrieve and manipulate data in a database.

#### DQL is used to query the database to extract specific data, and the most common DQL statement is the SELECT statement.

#### SELECT is used to retrieve data from a database table. It allows you to specify which columns you want to retrieve, from which table(s), and under which conditions.

#### example:

#### SELECT name, email
#### FROM customers
#### WHERE country = 'USA';


### ANSWER TO QUESTION 4:

#### Explain Primary Key and Foreign Key.

#### Primary Key (PK):

#### A Primary Key is a unique identifier for a record in a table. It ensures that each record has a unique value, preventing duplicates. A Primary Key:

#### 1. Uniquely identifies each record
#### 2. Must contain unique values
#### 3. Cannot be null
#### 4. Is used to reference related records in other tables

#### Example: In a "customers" table, "customer_id" could be the Primary Key, ensuring each customer has a unique identifier.

#### Foreign Key (FK):

#### A Foreign Key is a field in a table that references the Primary Key of another table. It establishes relationships between tables. A Foreign Key:

#### 1. Links to the Primary Key of another table
#### 2. Can contain null values
#### 3. Ensures consistency in relationships between tables

#### Example: In an "orders" table, "customer_id" would be a Foreign Key referencing the "customer_id" Primary Key in the "customers" table, linking orders to their respective customers.
